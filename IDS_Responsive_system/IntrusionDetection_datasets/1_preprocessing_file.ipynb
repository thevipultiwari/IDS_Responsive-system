<h6 style="font-family:san-serif;font-size:20px;color:black;font-weight:bold;text-align:left">Label Encoding</h6>

```python
protocol_dict={}
for idx,val in enumerate(data["protocol_type"].unique().tolist()):
    protocol_dict[val]=idx
print(protocol_dict)
```

```python
data["protocol_type"]=data["protocol_type"].map(protocol_dict)
```

```python
data.head()
```

```python
xattack_dict={}
for idx,val in enumerate(data["xAttack"].unique().tolist()):
    xattack_dict[val]=idx
print(xattack_dict)
```

```python
data["xAttack"]=data["xAttack"].map(xattack_dict)
```

```python
data.head()
```

```python
data.info()
```

<h6 style="font-family:san-serif;font-size:20px;color:black;font-weight:bold;text-align:left">Data resampling</h6>

```python
data["xAttack"].value_counts()
```

```python
from sklearn.utils import resample
resampled=[]
for label in data['xAttack'].unique().tolist():
    oversampled=resample(data.loc[data['xAttack']==label],n_samples=5000)
    resampled.append(oversampled)
df=pd.concat(objs=resampled,ignore_index=True)
df=df.sample(frac=1).reset_index(drop=True)
```

```python
df["xAttack"].value_counts()
```

```python
with plt.style.context(style='ggplot'):
    viz_data=dict(df['xAttack'].value_counts())
    fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(18,8))
    axes[0].bar(x=list(viz_data.keys()),
                height=list(viz_data.values()))
    axes[0].set_title(label='Analysing target feature using bar-chart')
    axes[0].set_xlabel(xlabel='class labels')
    axes[0].set_ylabel(ylabel='number of records')
    axes[1].pie(x=list(viz_data.values()),
                labels=list(viz_data.keys()),
                autopct='%.2f%%',pctdistance=0.60,
               startangle=90,
               shadow=True)
    axes[1].set_title(label='Analysing target feature using pie-chart')
    plt.show()
```

<h6 style="font-family:san-serif;font-size:20px;color:black;font-weight:bold;text-align:left">Data Normalization</h6>

```python
df.head()
```

```python
X=df.drop(labels='xAttack',axis=1)
X.head()
```

```python
y=df[['xAttack']]
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler=scaler.fit(X.values)
scaled_X=scaler.transform(X.values)
df=pd.DataFrame(data=scaled_X,columns=X.columns)
df['xAttack']=y
```

```python
df.head()
```

```python
import pickle
with open(file='models/Scaler.pkl',mode='wb') as file:
    pickle.dump(obj=scaler,file=file)
```

<h6 style="font-family:san-serif;font-size:20px;color:black;font-weight:bold;text-align:left">Data Splitting</h6>

```python
X=df.drop(labels='xAttack',axis=1)
X.head()
```

```python
y=df[['xAttack']]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True,random_state=42,stratify=y)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
X_train.to_csv('splitted_data/X_train.csv',index=False)
X_test.to_csv('splitted_data/X_test.csv',index=False)
y_train.to_csv('splitted_data/y_train.csv',index=False)
y_test.to_csv('splitted_data/y_test.csv',index=False)
```
